## Интерпретация полей

### 'bureau_AMT_CREDIT_SUM_min',
Может быть NaN

Current credit amount for the Credit Bureau credit

Сумма открытого кредита, флоат, может быть 0

Описательная статистика, минимум по аппликанту.

### 'prev_SELLERPLACE_AREA_min',
Может быть NaN

Selling area of seller place of the previous application

Торговая площадь места продавца (вероятно, если аппликантом был продавец). -1 - считать что такой площади нет, в остальном - положительное значение, флоат, может быть 0.

Описательная статистика, минимум по аппликанту.

### 'AMT_INCOME_TOTAL', 
Income of the client

Флоат в сотнях тысяч валюты, 1 знак после запятой, положительное число.

### 'prev_AMT_DOWN_PAYMENT_mean',
Может быть NaN

Down payment on the previous application

Размер авансового платежа по предыдущему кредиту, положительный флоат, может быть 0.

Описательная статистика, максимум по аппликанту.

### 'prev_WEEKDAY_APPR_PROCESS_START_SATURDAY_count_norm',
Может быть NaN

On which day of the week did the client apply for previous application

Может быть NaN

Признак, приходил ли клиент на оформление кредита в субботу (шолом шаббат), положительный флоат, может быть 0
Описательная статистика, среднее значение случившихся фактов по аппликанту.

### 'bureau_AMT_CREDIT_SUM_DEBT_max', 
Может быть NaN

Current debt on Credit Bureau credit

Текущая сумма открытого кредита, может быть отрицательной (вероятно, возобновляемый кредит, переплата), может быть 0

Описательная статистика, максимум по аппликанту.


### 'bureau_DAYS_CREDIT_UPDATE_mean',
Может быть NaN

How many days before loan application did last information about the Credit Bureau credit come

Количество дней до подачи заявки, за которое происходило последнее обновление по кредитам аппликанта, может быть положительной (допришедшие данные после подачи заявки) может быть 0.

Описательная статистика, среднее по аппликанту.


### 'bureau_DAYS_CREDIT_UPDATE_max',
Может быть NaN

How many days before loan application did last information about the Credit Bureau credit come

Количество дней до подачи заявки, за которое происходило последнее обновление по кредитам аппликанта, может быть положительной (допришедшие данные после подачи заявки) может быть 0.

Описательная статистика, максимум по аппликанту.

### 'prev_NAME_GOODS_CATEGORY_Furniture_count_norm',
Может быть NaN

What kind of goods did the client apply for in the previous application

Признак, что клиент покупал именно мебель на прошлые кредиты, положительный флоат, может быть 0

Описательная статистика, среднее значение случившихся фактов по аппликанту.


### bureau_DAYS_ENDDATE_FACT_max',
Может быть NaN

Days since CB credit ended at the time of application in Home Credit (only for closed credit)

Количество дней, прошедших с момента закрытия предыдущего кредита, отрицательное число (принятый формат от текущей даты), флоат, может быть 0

Описательная статистика, максимум по аппликанту.

### 'AMT_REQ_CREDIT_BUREAU_QRT',
Number of enquiries to Credit Bureau about the client 3 month before application

Количество обращений клиента в кредитное бюро за квартал до месяца, предшествующего подаче заявки, положительный флоат, может быть 0

### 'OWN_CAR_AGE',
